```python
# Cell 2 — Download dataset from Roboflow
from roboflow import Roboflow

rf = Roboflow(api_key="v3D73wkkizRLSXXOjIRn")
project = rf.workspace("sanjana-n9h7y").project("skinova-food-ostus")
version = project.version(1)
dataset = version.download("yolov11")
```

```python
# Cell 3 — Train YOLOv11
from ultralytics import YOLO

model = YOLO("yolo11m.pt")  # medium size — good balance of speed and accuracy

model.train(
    data=f"{dataset.location}/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    name="skinova_food_yolo",
    device=0,  # GPU
    patience=10,
    save=True
)
```

```python
# Cell 4 — Evaluate model
metrics = model.val()
print("mAP50:", metrics.box.map50)
print("mAP50-95:", metrics.box.map)
```
